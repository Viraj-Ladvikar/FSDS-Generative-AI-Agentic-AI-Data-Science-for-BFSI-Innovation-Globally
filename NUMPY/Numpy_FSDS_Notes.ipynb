## âœ… NumPy Arrays vs Python Sequences (Simple Points)

### ðŸ”¹ Size
- NumPy arrays have a **fixed size**
- Python lists can **grow or shrink anytime**

### ðŸ”¹ Memory Handling
- Resizing a NumPy array creates a **new array**
- The old array is discarded

### ðŸ”¹ Data Types
- NumPy arrays allow **only one data type**
- Python lists allow **mixed data types**

### ðŸ”¹ Performance
- NumPy arrays use **continuous memory blocks**
- This makes them **faster and more memory-efficient**

### ðŸ”¹ Mathematical Operations
- NumPy supports **fast vectorized math operations**
- Python lists are **slow for numerical computation**

### ðŸ”¹ Industry Usage
- Scientific and ML libraries **prefer NumPy arrays**
- Python lists are **automatically converted** to NumPy arrays internally

## Boolean Indexing 
- It allows you to select elements from an array based on a Boolean condition. This allows you
to extract only the elements of an array that meet a certain condition, making it easy to perform
operations on specific subsets of data.

```python
G = np.random.randint(1,100,24).reshape(6,4)
```

```python
G
```

```python
# find all the number graeter than 50 
G>50
```

```python
# Where is True , it gives result , everything other that removed.we got value

G[G>50]  # it is best Techinque to filter the data in given condition
```

```python
# find out the even numbera 

G[G%2==0]
```

```python
# find all the number greater than 50 and evena 
G[(G>50) & (G%2==0)]  # Here we used (&) bitwise Not logical(and) , because we are working with boolean values
```

```python
# find all number not divisible by 7
~(G%7==0)
```

```python
G[~(G%7==0)]
```

## BroadCasting 

- Used in Vectorization
-The term broadcasting describes how NumPy treats arrays with different shapes during
arithmetic operations.
- The smaller array is â€œbroadcastâ€ across the larger array so that they have compatible shapes.

```python
# same shapea 
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)
print()
print(a+b)
```

```python
# diff shape 
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print()

print(a+b)
```

```python
# more examplea 
a = np.arange(12).reshape(4,3)

b = np.arange(3)

print(a)
print(b)
```

```python
print(a+b ) #Arthematic Operation possible because , Here a = (4,3) is 2D and b =(3) is 1D so did converted (3) to (1,3) and streched to (4,3)
```

```python
# Could not Broadcast
a = np.arange(12).reshape(3,4)
b = np.arange(3)
print(a)
print(b)
print(a+b)
# Arthematic Operation not possible because , Here a = (3,4) is 2D and b =(3)
#is 1D so did converted (3) to (1,3) and streched to (3,3) but , a is not equals to b . so it got failed
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)
print(a)
print(b)
print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)
print(a)
print(b)
print(a + b)
```

```python
a = np.array([1])
# shape -> (1,1) streched to 2,2
b = np.arange(4).reshape(2,2)
# shape -> (2,2)
print(a)
print(b)
print(a+b)
```

```python
# doesnt work
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)
print(a)
print(b)
print(a+b)
```
