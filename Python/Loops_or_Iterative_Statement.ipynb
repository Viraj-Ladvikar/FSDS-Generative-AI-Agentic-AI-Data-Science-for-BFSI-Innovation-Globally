## 1) For Loop
 - For loops is used to iterate over a sequence such as a list, tuple, string or range. It allow to execute a block of code repeatedly, once for each      item in the sequence.

```python
for x in range(0,5):
    print(x)
```

## While Loop
- if we wnat to execute a group of statement iteratvily until some conditoin false 

```python
i = 5 

while i>=0:
    print("Data Scinece")
    i-=1
```

```python
i = 0 

while i <=5:
    print("FSDS")
    i+=1
```

```python
i = 1 

while i<=5:
    print('data science') # when we mention end then new line will not create
    j = 1
    
    while j<=4:
        print('technology')
        j = j + 1
        
    i = i + 1
    print()
```

```python
i = 1 

while i<=5:
    print('data science',end=" ") # when we mention end then new line will not create
    j = 1
    
    while j<=4:
        print('technology',end=" ")
        j = j + 1
        
    i = i + 1
    print()
    
```

```python
# lets use while loop usig some numbers
i = 1
while i <= 4 :
    j = 0
    while  j <= 3 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

```python
i = 1

while i <= 2 :
    j = 0
    while  j <= 2 :
        print(i*j, end=" ")
        j += 1
    print()
    i += 1
```

## Loop Control Statement

### Continue Statement
- With the continue statement we can stop the current iteration of the loop, and continue with the next:
- or Skipped value 

```python
for i in range(0,11):
    if i%2==0:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

## Break Statement
 - The break statement in Python brings control out of the loop.

```python
for x in range(0,10):
    if x == 5:
        break
    print(x)
```

```python
for i in "Viraj":

    if i == "a":
        break
    print(i)
```
