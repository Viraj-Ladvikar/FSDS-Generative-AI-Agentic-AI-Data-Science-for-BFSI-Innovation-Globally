# Type Casting

- We can convert one type value to another type is called typecasting

- ### Inbuilt Function for type casting
  1) int()
  2) float()
  3) complex()
  4) str()
  5) bool()

### 1) int() : convert values from other  type to int 
  - we can convert any other type value to int type except complex and text string  

```python
a = int(3.3)
b = int(3)
c = int("10")
d = int(True)
e = int(True+True)
print(a,type(a))
print(b,type(b))
print(c,type(c)) 
print(d,type(d))  
print(e,type(e))  
```

```python
a  = int(2+3j)
```
