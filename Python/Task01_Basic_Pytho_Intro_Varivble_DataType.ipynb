### Int Data Type
 - to represent whole number (integral vlaue)

```python
a = 10 # int
print(a,type(a))
```

#### We can Represent Int value in following way
 1) Decimal form
 2) Binary formm
 3) octal form
 4) hexa decimal form

#### 1) Decimal from
 - allowed digit in between 0-9

```python
a = 9
print(a,type(a))
```

#### 2) Binary Form
 - allowed digit  0 & 1
 - prefixed with 0b or 0B

```python
a = 0b111
print(a,type(a))
```

```python
b = 0B0101
print(b ,type(b))
```

#### 3) Octal form
 - allowed digit 0-7
 - prefixed with 0o or 0O

```python
a = 0o123
print(a,type(a))
```

#### 4) Hexa Decimal Form
 - allowed digit 0- 9 ,a-f(bothn upper and lower ) a=10,b=11,c=12,d=13,e=14,f=15
 - prefixed with 0x or 0X

```python
a = 0xAFEC
print(a)
```

```python
b = 0XBEEF
print(b)
```

#### Base Conversion

##### 1) bin()
 - we can use bin() to convert any base to binary

```python
a = bin(16)
print(a)
```

```python
b = bin(0b1111)
print(b)
```

##### 2) oct()
 - we canuse oct() to convert from any base to octal 

```python
a = oct(19)
print(a)
```

```python
b = 0O1233
print(b)
```

##### 3) hex()
 - we can use hex() to convert from any base to hexa decimal

```python
a = hex(123)
print(a)
```
