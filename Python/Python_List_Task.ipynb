## How to access the element from the list and what are the ways ?
 - Using Index [Index]
 - Using Slice  [Start : End : Step]

```python
l = [10,20,30,40]
print(l[3]) # 40 using index
print(l[0:2]) # using Slice 
print(l[0:])
print(l[:7])
print(l[::-1]) 
print(l[-1:-3:-1])
#print(l[8])  #IndexError: list index out of range 
```

## ALL/ANY
 - âœ… any()

ðŸ”¹ Returns True if at least one element in an iterable (list, tuple, etc.) is True
ðŸ”¹ Returns False only when all elements are False / empty / 0 / None

 - ðŸ”¹ all()

ðŸ”¹ Returns True when every element in an iterable is True
ðŸ”¹ Returns False if any one element is False 

```python
l = [1,2,3,0]
print(any(l))
print(all(l))
```

```python
l2 = [1,2,2,2]
print(any(l2))
print(all(l2))
```

```python
l3 = [0,False,""]
print(any(l3))
print(all(l3))
```
