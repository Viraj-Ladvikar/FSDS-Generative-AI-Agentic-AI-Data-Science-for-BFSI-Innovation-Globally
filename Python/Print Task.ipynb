# Print Format Method

```python
num1=20
num2=30
add = num1+num2
print("the addition of {} and {} is = {}".format(num1,num2,add))
```

```python
num1=100
num2=25
num3=333
avg=(num1+num2+num3)/3
avg1=round((num1+num2+num3)/3,2)
print('the average of {} ,{} and {} is  ={} or{}'.format(num1,num2,num3,avg,avg1))
```
