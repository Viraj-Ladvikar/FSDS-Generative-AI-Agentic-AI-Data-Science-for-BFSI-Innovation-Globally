# Condiitonal Statement
 - Conditional statements in Python are used to execute certain blocks of code based on specific conditions.
 - Types of Condiitonal Statement
     - if
     - if else
     - if elif else

## If Conditional Statement
  ### Synatx :
   - if condition:
        Statement

```python
age = 18

if age >=18:
    print("Eligible")
```

```python
name = input("Enter Your Name")

if name =="Viraj":
    print(f'Hi, {name}!')
```

```python
x = 5
r = x % 2 

if r == 0:
    print('Even number')
    
print('odd number')
```

```python
x = 3

if x%2==0:
    print("even")
if x%2!=0:
    print("odd")
```

```python
# Short Hand If

a = 10

if a%2==0:print("even")

```

### If else Conditional Statement
 - if the condition is true then if block execute other wise else block

```python
age = int(input("Enter your age "))

if age>18:
    print("Eligible")
else:
    print("not Eligible")
```

```python
# Short Hand if eFalse

age  = int(input("Enter your age "))

print("Eligible") if age>=18  else print("not eligivle")
```

```python
x = 3
r = x % 2

if r == 0:
    print('Even number')
    if x>5:
        print('greater number')
  
else:
    print('Odd Number')
```

### Nested if else

```python
x = 4
r = x % 2

if r == 0:
    print('Even number')
    
    if x>5:
        print('greater number')
  
else:
    print('Odd Number')
```

### elif Statement


```python
x = int(input("Enter number"))

if x == 1:
    print('one')
elif x == 2:
    print('Two')
elif x == 3:
    print('Three')
elif x == 4:
    print('four')
else:
    print("invalid")
```

## Q1) WAP to find biggest of given 2 number from the command prompt ?

### Ternary Conditional Statement


```python
# Assign a value based on a condition
age = 20
s = "Adult" if age >= 18 else "Minor"
print(s)
```
