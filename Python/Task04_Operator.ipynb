### Assingement operator 

```python
a = 10

a += 1
print(a)   # 11

a -= 1
print(a)   # 10

a *= 2
print(a)   # 20

a /= 5
print(a)   # 4.0  (float!)

a //= 2
print(a)   # 2.0  (still float)

print(a ** 2)  # 4.0
```
