# Dictionary dict
  - Key:value pair
  - Duplicate keys not allowed but duplicate value allowed
  - insertion order not maintain
  - mutable
  - indexing and slicing not allowed
  - heterogenous data allowed

## Dictionary Creation 

```python
d ={}
print(type(d))
print(d)
```

```python
d ={1:"ones",2:"two",3:"Three"}
print(type(d))
print(d)
```

```python
d = dict({1:11,2:22,3:33})
print(d)
print(type(d))
```

```python
d = {"ones":1,"two":2}
print(d)
print(type(d))
```

## copy():- Return a shallow copy of the dict

```python
mydict1 = {'Name':'Asif' , 'ID': 74123 , 'DOB': 1991 , 'job' :'Analyst'} 
mydict1
```

```python
mydict2 = mydict1.copy()
print(mydict2)
print(id(mydict1)==id(mydict2))
```

```python
mydict3 =mydict1  # create a anew reference

print(id(mydict3),id(mydict1))
```

## Acces item using loop 

```python
mydict1 = {'Name':'Asif' , 'ID': 74123 , 'DOB': 1991 , 'job' :'Analyst'} 
mydict1
```

```python
for i in mydict1:
    print(i) # only keys print
    print(f"{i} : {mydict1[i]}")
```
