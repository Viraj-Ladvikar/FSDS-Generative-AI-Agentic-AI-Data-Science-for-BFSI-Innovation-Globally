# Website Analysis

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv(r"C:\Users\ladvi\Downloads\data-export (1).csv")
```

```python
df
```

```python
df.columns = df.loc[0]
df = df.drop(index=0).reset_index(drop=True)
df.columns = [ "channel group ",	"DateHour",	"Users",	"Sessions"	,"Engaged sessions",	"Average engagement time per session","Engaged sessions per user"	,"Events per session",	"Engagement rate",	"Event count"]
```

```python
df
```

```python
df.info()
```

```python
df['DateHour'] = pd.to_datetime(
    df['DateHour'].astype(str),
    format='%Y%m%d%H',
    errors='coerce'
)
```

```python
df['Hour']  = df['DateHour'].dt.hour
```

```python
df.head()
```

```python
df.rename(columns={"channel group ":'channel group'},inplace=True)
```

```python
numeric_cols = df.columns.drop(["channel group","DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric,errors='coerce')
```

# What pattern or trends can you observe in website session and user over time?

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Session and user over time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

# Which marketing channel brought the heighest number of user to the website , and how can we use this insight to improve traffice from other ?

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='channel group',y='Users',estimator=np.sum,palette="viridis")
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

# which channel has the height average engagement time and what does that tell us about user behaviour and content diffrence 

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='channel group',y='Average engagement time per session',estimator=np.mean,palette="magma")
plt.title("Average Engagement Time By Channel")
plt.xticks(rotation=45)
plt.show()
```

# How does engagement rate varry across different trafficx channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,y="Engagement rate",x="channel group",palette="coolwarm")
plt.title("Engagemnt rate distribution channel")
plt.xticks(rotation=45)
plt.show()
```

# Engaged vs non-engaged session

```python
df.head()
```

```python
session_df = df.groupby("channel group") [["Sessions","Engaged sessions"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"]-session_df["Engaged sessions"]
session_df_melted = session_df.melt(id_vars="channel group",value_vars=["Engaged sessions","Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted,x="channel group",y="value",hue="variable")
plt.title("Enaged vs non engaged")
plt.xticks(rotation=45)
plt.show()
```
