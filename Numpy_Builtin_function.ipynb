# Numpy Built-in Function

```python
import numpy as np
```

## Array Creation

```python
# Create an array from llist

arr = np.array([1,2,3,4])
arr
```

```python
# create an array with evenly spaced value

b = np.arange(0,10,2)
print(b)

c = np.arange(-1,-20,-3)
print(c)
```

```python
# Create an array filled with zero

d = np.zeros((2,3))   # here 2 * 3 is shape 
print(d)
```

```python
# Create an array filled with oneabs 

e = np.ones((3,4))
print(e)
```

```python
# Create an identity matrix 
f = np.eye(5)  # 5*5 matrix 
print(f)
print("================")

g = np.eye(3,4)
print(g)

print("================")
h = np.eye(3,4,k=1)   # diagonnal shifet one above 
print(h)
print("================")
i = np.eye(3,4,k=-1)    #  diagonal shifted one belowe
print(i)
print("================")
```

## 2. Array Manipulation Function

```python
# Reshape an array 

a1= np.array([1,2,3,4])
print(a1)

print(a1.shape)

print()

a1 = a1.reshape(2,2)
print(a1)

print(a1.shape)
```

```python
# flatten an array

a = np.array([[1,2],[3,4]])
print(a)
print()
a_ravel  = np.ravel(a)
print(a_ravel)
print(a_ravel.shape)
print()
a_flatten = a.flatten()
print(a_flatten)
print(a_flatten.shape)

print("======================")

a[0,0] = 32

print(a)
print("======================")
print(a_ravel)
print("======================")
print(a_flatten)
```

```python
# Transpose of matrix
t1 = np.array([[1,2],[3,4]])
print(t1)
print('=============================')
print(t1.T)
```

```python
# Stack arrays verticallya
a = np.array([1,2,3])
b = np.array([4,5,6])

v_stacked = np.vstack((a,b))
print(v_stacked)
```

```python
# stack array horizontally

a = np.array([1,2,3])
b = np.array([4,5,6])

h_stacked = np.hstack((a,b))
print(h_stacked)
```

```python
# stack array with hlp of axis 
a = np.array([1,2,3])
b = np.array([4,5,6])

h_stacked = np.stack((a,b),axis=1)
v_stacked = np.stack((a,b),axis=0)
print(h_stacked)
print(v_stacked)
```

# # 6. Random Sampling Functions

```python
# Generate random values between 0 and 1
random_vals = np.random.rand(3)  # Array of 3 random values between 0 and 1
print("Random values:", random_vals)
```

```python
# Generate random values between 0 and 1
random_vals = np.random.randint(3)  # Array of 3 random values between 0 and 1
print("Random values:", random_vals)
```

```python
# Set seed for reproducibility
np.random.seed(0)

# Generate random integers
rand_ints = np.random.randint(0, 10, size=5)  # Random integers between 0 and 10
print("Random integers:", rand_ints)
```

# 7. Boolean & Logical Fucntion 

```python
# Check if all elements are True
# all
logical_test = np.array([True, False, True])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
logical_test = np.array([True, True, True])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
logical_test = np.array([False, False, False])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
# Any : check if any of the element are ture

logical_test = np.array([False,True,False])
any_true = np.any(logical_test)
print("Any elements True:", any_true)
```

# 8. Set Operator

```python
#Intersection of two arrays
set_a = np.array([1,2,3,4])
set_b= np.array([3,4,5,6])

intersection = np.intersect1d(set_a,set_b)
print(intersection)
```

```python
# Union of two array 

union = np.union1d(set_a,set_b)
print(union)
```

# 9. Array Attribute an function 

```python
# Array attributes
a = np.array([1, 2, 3])
shape = a.shape  # Shape of the array
size = a.size    # Number of elements
dimensions = a.ndim  # Number of dimensions
dtype = a.dtype   # Data type of the array
itemsize = a.itemsize
print("Shape of a:", shape)
print("Size of a:", size)
print("Number of dimensions of a:", dimensions)
print("Data type of a:", dtype)
print("Itemsize type of a:", itemsize)
```
