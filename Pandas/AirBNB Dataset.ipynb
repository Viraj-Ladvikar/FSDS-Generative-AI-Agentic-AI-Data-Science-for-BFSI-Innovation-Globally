# Handling Mising values

```python
airbnb['last review'] = pd.to_datetime(airbnb['last review'],errors='coerce')   # errror will convert any nan missing value in nat (not a think ) value
```

```python
airbnb.info()
```

```python
airbnb.fillna({'reviews per month':0,'last review':airbnb['last review'].min()},inplace=True)
```

```python
airbnb.dropna(subset=['NAME','host name'],inplace=True)
```

```python
print(airbnb.isnull().sum())
```

```python
airbnb = airbnb.drop(columns=['house_rules','license'],errors='ignore')
```

```python
airbnb.head()
```

```python
airbnb['price']  = airbnb['price'].replace('[\$,]','',regex=True).astype(float)
airbnb['service fee'] = airbnb['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
airbnb.info()
```

```python
airbnb.head()
```

# Remove Duplicates

```python
airbnb.drop_duplicates(inplace=True)
```

```python
airbnb.info()
```

# Visualization

## What is the distribution of lisitng prices

```python
plt.figure(figsize=(10,6))
sns.histplot(airbnb['price'],bins=50,kde=True,color='red')
plt.title("Distribution of listing Price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

## How are diffrent rooms types distributed

```python
airbnb['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='room type',data=airbnb,color='hotpink')
plt.title("room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()
```

# How are lisiting distributed across differnt neighbourhood

```python
airbnb['neighbourhood']
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group',data=airbnb,color="teal",order=airbnb['neighbourhood group'].value_counts().index)
plt.xlabel("Neighbourhood")
plt.ylabel("Count")
plt.title("Neighbourhood Distribution")
plt.show()
```

## What is the relationshiop between price and room type

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='room type',y='price',hue='room type',data=airbnb,palette='Set1')
plt.xlabel("RoomType")
plt.ylabel('Price ($)')
plt.title("Price Vs Room Type")
plt.legend(title='Rooom Type')
plt.show()
```

# How has the number of review change over with time

```python
airbnb.info()
```

```python
airbnb['last review'] = pd.to_datetime(airbnb['last review'])
review_over_time = airbnb.groupby(airbnb['last review'].dt.to_period(('M'))).size()
plt.figure(figsize=(12,6))
review_over_time.plot(kind='line',color='red')
plt.title("Number of review overetime")
plt.xlabel("Date")
plt.ylabel("Number of review")
plt.show()
```
