```python
import matplotlib.pyplot as plt
import seaborn as sns



%matplotlib inline
plt.rcParams['figure.figsize'] = 6,3

import warnings

warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
vis1 =  sns.distplot(df["InternetUsers"]) 
plt.show()
```

```python
vis2 =  sns.displot(df["InternetUsers"]) 
plt.show()
```

```python
vis1 =  sns.histplot(df["InternetUsers"],bins=15) 
plt.show()
```

```python
vis1 =  sns.histplot(df["InternetUsers"]) 
plt.show()
```

```python
df
```

```python
plt.rcParams['figure.figsize'] = 9,5
vis4 =  sns.boxplot(data = df,x="IncomeGroup",y="BirthRate") 
plt.show()
# outlier is  data point which is very far fromother observation 
# also called as anomoly detecion
```

```python
vis5 = sns.lmplot(data = df,x="InternetUsers",y="BirthRate")
plt.show()
```

```python
vis5 = sns.lmplot(data = df,x="InternetUsers",y="BirthRate",fit_reg=True)
plt.show()
```

```python
vis6 = sns.lmplot(data = df,x="InternetUsers",y="BirthRate",fit_reg=True,hue="IncomeGroup")
plt.show()
```

```python
vis6 = sns.lmplot(data = df,x="InternetUsers",y="BirthRate",fit_reg=False,hue="IncomeGroup")
plt.show() # hue act liek legend
```
