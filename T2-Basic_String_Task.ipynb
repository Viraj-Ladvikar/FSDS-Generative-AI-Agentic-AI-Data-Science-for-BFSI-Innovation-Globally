# Slicing

```python
language ="Python"
print(len(language))
first_three = language[0:3]
print(first_three)
last_three = language[3:6]
print(last_three)
first_three2= language[:3]
print(first_three2)
last_three2 =language[3:]
print(last_three2)
last_three3 = language[-3:]
print(last_three3)
```

### expandtabs(): Replaces tab character with spaces, default tab size is 8. It takes tab size argument

```python
language = "Hello\tworld\tpython"
print(language)
print(language.expandtabs())
print(language.expandtabs(10))
print(language.expandtabs(0))
```
